```python
matrix= [[0 for x in range(3)] for y in range(3)] 
player=1
moveCount=0
def printBoard():
    toPrint=""
    for x in range(len(matrix)):
        toPrint+=" \n "
        for y in range(len(matrix[x])):
            if matrix[x][y]==0:
                toPrint+="|-|"
            if matrix[x][y]==1:
                toPrint+="|1|"
            if matrix[x][y]==2:
                toPrint+="|2|"
    print toPrint
    pass
    
def startup():
    global matrix
    matrix= [[0 for x in range(3)] for y in range(3)] 
    global player,moveCount
    player=1
    moveCount=0
    while True:
        printBoard()
        x=input("Enter x position:")
        y=input("Enter y position:")
        if x<0 or y<0 or x>2 or y>2:
            print "Enter valid position"
            continue
        if matrix[x][y]!=0:
            print "This position is taken"
            continue
        if player==1:
            matrix[x][y]=1
            player=2
        else:
            matrix[x][y]=2
            player=1
        win=checkWinner(x,y,1 if player==2 else 2)
        if win==3:
            print "No winner."
            startup()
        if win!=0:
            print "Play again:",win
            startup()
        else:
            continue
    pass
    
def check(count,play):
    if count==3:
        print "Player %d"%play+" won"
        printBoard()
        return True
    return False
       
def checkWinner(x,y,play):
    count=0
    global moveCount
    moveCount+=1
    length=len(matrix)
    #cols
    for j in range(length):
        if matrix[x][j]==play:
            count+=1
    if check(count,play):
        return play
    #rows
    count=0
    for i in range(length):
        if matrix[i][y]==play:
            count+=1
    if check(count,play):
        return play
    #diagonal
    count=0
    for i in range(length):
        if matrix[i][i]==play:
            count+=1
    if check(count,play):
        return play
    # anti-diagonal
    count=0
    for i in range(length):
        if matrix[i][length-i-1]==play:
            count+=1
    if check(count,play):
        return play
    if moveCount==9:
        return 3
    return 0
    
```
